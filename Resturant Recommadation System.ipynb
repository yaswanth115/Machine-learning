```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("Resturants dataset.csv")
df
```

```python
df['Cuisines'].fillna('Unknown', inplace=True)
print(df.isnull().sum())
```

```python
features = df[['Restaurant Name', 'Cuisines', 'Price range', 'Aggregate rating']]
```

```python
# One-hot encoding for Cuisines
encoder = OneHotEncoder()
cuisines_encoded = encoder.fit_transform(features[['Cuisines']]).toarray()
# Normalize numerical features
scaler = MinMaxScaler()
price_scaled = scaler.fit_transform(features[['Price range']])
rating_scaled = scaler.fit_transform(features[['Aggregate rating']])

# Combine all features into a single matrix

combined_features = np.hstack([cuisines_encoded, price_scaled, rating_scaled])
```

```python
scaler = MinMaxScaler()
encoder = OneHotEncoder()

# Example user preferences
user_preference = {'Cuisines': 'Italian', 'Price range': 3, 'Aggregate rating': 4.5}

# Create DataFrame for user preferences with the same column names as used for fitting
scaler.fit(df[['Price range', 'Aggregate rating']])

# Now transform the user data using both 'Price range' and 'Aggregate rating'
user_data = pd.DataFrame(
    [[user_preference['Price range'], user_preference['Aggregate rating']]], 
    columns=['Price range', 'Aggregate rating']
)

# Apply scaling
user_scaled = scaler.transform(user_data)

# Extract scaled values
scaled_price = user_scaled[0][0]
scaled_rating = user_scaled[0][0]

# For the 'Cuisines' feature, apply OneHotEncoding (assume you have the encoder from training)
user_cuisine_encoded = encoder.transform(pd.DataFrame([user_preference['Cuisines']], columns=['Cuisines'])).toarray()

# Combine all features (scaled price, scaled rating, and encoded cuisine) into a single array
user_features = np.hstack([user_cuisine_encoded, np.array([[scaled_price, scaled_rating]])])

print(user_features)
```
